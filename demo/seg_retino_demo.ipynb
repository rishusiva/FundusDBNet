# **Retina Blood Vessel Segmentation**

This is the standalone demo notebook to demonstrate Retinal Blood Vessel Segmentation using our pretrained UNET weights. The demo notebook uses images that come with the repo cloning itself, however you may use images of your choice once you change the path.

```python
from segRetino.inference import SegRetino

# Creating the SegRetino object initialized with the test image path
seg = SegRetino('/content/segRetino/results/input/input1.png')

# Running inference
seg.inference(set_weight_dir = 'unet.pth', path = '/content/output1.png', blend_path = '/content/blend1.png')
```

```python
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open('/content/segRetino/results/input/input1.png')
seg = Image.open('/content/output1.png')
blend = Image.open('/content/blend1.png')
plt.imshow(img)
plt.show()
plt.imshow(seg)
plt.show()
plt.imshow(blend)
plt.show()
```
